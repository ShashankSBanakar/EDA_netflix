# matplotlib.pyplot
> matplotlib has 2 modules: one for matlab background programmers and the other, pyplot for pyhton background


> canvas is the entire plot including the space beyond the axis

> xticks,yticks

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
x=np.arange(1,100)
y=x*2 + 3
z=x*3 - 2
```

```python
plt.style.use('seaborn')                                    # seaborn creates more visually attractive graphs, like gridlines
fig=plt.figure(figsize=(6,4))                               # change the figure size
plt.plot(x,y,label='First')
plt.plot(x,z,label='Second')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Demo Plot')
plt.legend()
plt.show()
```

# subplot
> dividing the canvas into 2 parts and plotting 2 graphs

```python
plt.style.use('seaborn')                          
# fig=plt.figure(figsize=(6,4))
fig, ax = plt.subplots(1,2,figsize=(8,6))                          # Subplot: row=1, columns=2, figure_size=(8,6). Dividing the canvas into 2 columns
                                                                   # plt.subplot gives 2 objects: a figure and axis: ax
                                                                   # ax[0]: is the first plot, ax[1]: second plot
                                                                   # individually alot the xlabel, ylabel and legends for ax[0] and ax[1]
    
ax[0].plot(x, y, '--r', label = 'Plot1', alpha=0.5)                # x_axis=x, y_axis=y, the contour:'--', r stands for red.
                                                                   # alpha: controls the transarency
ax[1].plot(x, z, '-.k', label = 'Plot2', linewidth=3)              # contour:'_.', k stands for black in pyplot, linewidth: thickness of the plot

ax[0].set_xlabel('X')
ax[0].set_ylabel('Y')

ax[1].set_xlabel('X')
ax[1].set_ylabel('Y')

ax[0].legend(loc=0)                                                 # change the position of the legend with 'loc'. 0<loc<10
ax[1].legend()
# plt.show()
```

```python
x = np.linspace(-10,10,1000)
y = np.sin(x)
z = np.cos(x)
```

```python
plt.style.use('seaborn')
fig = plt.figure(figsize=(12,8))

ax1 = fig.add_axes([0.1,0.1,0.4,0.7])
ax2 = fig.add_axes([0.6,0.1,0.4,0.2])
            
ax1.plot(x,y)
ax2.plot(x,z,color='purple')

plt.show()
```
